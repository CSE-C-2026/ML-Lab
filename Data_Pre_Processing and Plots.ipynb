# Plots (iris dataset)

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
data=pd.read_csv("iris.csv")
id = [i for i in range(150)]
data["id"] = id
data = data.reindex(columns=['id','sepal.length','sepal.width','petal.length','petal.width','variety'])
print(data)
```

```python
#Line plot
plt.figure(figsize=(4,4))
plt.plot(data["id"],data["sepal.length"])
plt.xlabel("ID")
plt.ylabel("Petal length")
plt.title("Line Plot ID vs Sepal length")
plt.show()
```

```python
#Line plot for 4 cols
plt.plot(data['sepal.width'],label="Sepal_Width",color="blue")
plt.plot(data['sepal.length'],label="Sepal_Length",color="red")
plt.plot(data['petal.width'],label="Petal_Width",color="green")
plt.plot(data['petal.length'],label="Petal_Length",color="purple")
plt.xlabel("Id")
plt.ylabel("cms")
plt.title("Line Plot for 4 columns")
plt.legend()
```

```python
#Scatter plot
x=data["sepal.length"]
y=data["petal.length"]
z=data["id"]
plt.figure(figsize=(4,4))
plt.scatter(z,x,color="blue",label="Sepal_Length")
plt.scatter(z,y,color="green",label="Petal_Length")
#Adding labels to the axes
plt.xlabel("Id")
plt.ylabel("cms")
#Adding title to the plot
plt.title("Scatter plot for Sepal Length and Petal Length")
#Adding legen to the axes
plt.legend()
#Displaying the plot
plt.show()
```

```python
#subplot for all four features (line charts)
id=data['id']
x=data['sepal.length']
y=data['petal.length']
z=data['sepal.width']
w=data['petal.width']
fig,axs = plt.subplots (2, 2, figsize=(5, 3))
axs[0,0].plot(id,x, color='red') #id vs Sepal Length
axs[0,0].set_xlabel('Id')
axs[0,0].set_ylabel('Petal Length')
axs[0,1].plot(id, y, color='blue') #id vs Petallength
axs[0,1].set_xlabel('Id')
axs[0,1].set_ylabel('Sepal width')
axs[1,0].plot(id, z, color='green') #id vs Sepalwidth
axs[1,0].set_xlabel('Id')
axs[1,0].set_ylabel('Petal Width')
axs[1,1].plot(id, w, color='orange') #id vs Petalwidth
axs[1,1].set_xlabel('Id')
axs[1,1].set_ylabel('Sepal Widtd')
#Adjust Layout
plt.tight_layout()
#Show plot
plt.show()
```

```python
#subplot for all four features (scatter plots)
id=data['id']
x=data['sepal.length']
y=data['petal.length']
z=data['sepal.width']
w=data['petal.width']
fig,axs = plt.subplots (2, 2, figsize=(5, 3))
axs[0,0].scatter(id,x, color='red') #id vs Sepal Length
axs[0,0].set_xlabel('Id')
axs[0,0].set_ylabel('Petal Length')
axs[0,1].scatter(id, y, color='blue') #id vs Petallength
axs[0,1].set_xlabel('Id')
axs[0,1].set_ylabel('Sepal width')
axs[1,0].scatter(id, z, color='green') #id vs Sepalwidth
axs[1,0].set_xlabel('Id')
axs[1,0].set_ylabel('Petal Width')
axs[1,1].scatter(id, w, color='orange') #id vs Petalwidth
axs[1,1].set_xlabel('Id')
axs[1,1].set_ylabel('Sepal Width')
#Adjust Layout
plt.tight_layout()
#Show plot
plt.show()
```

# Normalization

```python
#Importing MinMaxScaler and StandardScaler Library from sklearn.preprocessing
from sklearn.preprocessing import MinMaxScaler,StandardScaler
```

```python
#Assigning StandardScaler() to scaler
scaler=StandardScaler()
```

```python
#Plotting histogram for 'Purchase' before normalising
plt.figure(figsize=(4,4))
plt.hist(data['Purchase'],edgecolor="black")
plt.title("Histogram of Purchase")
plt.xlabel("Scale")
plt.ylabel("Frequency")
```

```python
#Plotting Histogram for Purchase after normalising
data1=data[['Purchase']]
data['Purchase']=scaler.fit_transform(data1)
plt.figure(figsize=(4,4))
plt.title("Histogram of Purchase")
plt.hist(data['Purchase'],edgecolor='black')
plt.xlabel("Scale")
plt.ylabel("Frequency")
```

```python
#Assigning MinMaxScaler to scaler
scaler=MinMaxScaler()
```

```python
#Plotting Histogram for 'Occupation' before Normalising using MinMaxScaler
plt.figure(figsize=(4,4))
plt.hist(data['Occupation'],edgecolor="black")
plt.title("Histogram of Occupation")
plt.xlabel("Scale")
plt.ylabel("Frequency")
```

```python
#Plotting Histogram for 'Occupation' after normalising
data2=data[['Occupation']]
data['Occupation']=scaler.fit_transform(data2)
plt.figure(figsize=(4,4))
plt.title("Histogram of Occupation")
plt.xlabel("Scale")
plt.ylabel("Frequency")
plt.hist(data['Occupation'],edgecolor="black")
```
